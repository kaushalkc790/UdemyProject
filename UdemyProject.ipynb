### Find out and display number of courses per subject

```python
udemy.groupby('subject')['course_title'].count()
```

```python
udemy['subject'].value_counts().reset_index()
```

```python
udemy['subject'].unique()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x=udemy['subject'])
plt.grid()
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(x=udemy['subject'],hue = udemy['level'])
for container in ax.containers:
    ax.bar_label(container)
plt.ylabel("Number of subscribers")
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(x=udemy['level'],hue = udemy['subject'])
for container in ax.containers:
    ax.bar_label(container)
plt.ylabel("Number of subscribers")
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(x=udemy['level'],hue = udemy['is_paid'])
for container in ax.containers:
    ax.bar_label(container)
plt.ylabel("Number of subscribers")
```

```python
subject_count= udemy['subject'].value_counts()
plt.pie(subject_count,autopct='%1.1f%%',labels=subject_count.index)
plt.show()
```

### Find the revenue by level

```python
udemy['revenue'] = udemy['price'] * udemy['num_subscribers']
udemy.head()
```

```python
udemy.groupby(['level','subject'])['revenue'].sum()
```

```python
udemy.groupby(['level'])['revenue'].sum()
```

### Highest subscribers in particular year

```python
udemy.head()
```

```python
udemy['year'] = udemy['published_timestamp'].dt.year
udemy.groupby('year')['num_subscribers'].sum().sort_values(ascending=False)
```

```python
udemy.groupby('year')['num_subscribers'].sum().sort_values(ascending=False).idxmax()
```

```python
udemy.groupby('year')['num_subscribers'].max()
```

```python
idx_max = udemy.groupby('year')['num_subscribers'].idxmax()
udemy.loc[idx_max,['num_subscribers','year','course_title','subject']]
```

```python
udemy.loc[idx_max]
```

```python
udemy.groupby(['year','level'])['num_subscribers'].idxmax().reset_index(name='idx_max_subscribers')
```

```python
idx=udemy.groupby(['year','level'])['num_subscribers'].idxmax().reset_index()['num_subscribers']
udemy.loc[idx].sort_values(by=['year','level'])[['year','level','course_title','num_subscribers']]
```

```python
udemy[udemy['year']==2011]
```

```python
udemy.groupby(['year','subject'])[['revenue']].sum()
```

```python
udemy.groupby(['year','subject'])[['revenue']].sum().plot(kind='bar')
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(udemy,x='year', y='revenue',estimator = max,hue='subject') #default mean revenue
for container in ax.containers:
    ax.bar_label(container)
```
